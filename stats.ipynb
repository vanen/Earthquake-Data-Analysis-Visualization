```python
%matplotlib inline
# add necessary import statements
import sys
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ADD input validation to make sure an integer between 1 and 20 is provided
# Program should repeat request for a number OR exit if invalid input is provided
def validate_user_input(howmany, min, max):
    try:
        howmany = int(howmany)
        if(howmany >= min and howmany <= max):
            return howmany
        return 0
        print("Value out of bounds.")
    except ValueError:
        sys.exit("You have entered an invalid input. ")

while True:
    howmany = input(("How many earthquake locations to show? "))
    howmany = validate_user_input(howmany, 1, 20)
    if howmany > 0: break
    
# connect to the output database and name it earthquakes.sqlite 
conn = sqlite3.connect('earthquakes.sqlite')

# forces database to return strings for TEXT attributes 
conn.text_factory = str 
# get the cursor for the connection 
cur = conn.cursor()

# run select query on database to get data for statistics
cur.execute('''SELECT magnitude, time, depth, place, title, felt, region 
           FROM earthquakes''')

# set up variables to hold database data
# will be counting earthquakes by place/region
# will just be storing data about magnitude, felt and tsunami 
placecounts = dict()
regioncounts = dict()
mag = []
felt = []
#tsunami = []  # Per the last assignment, this was not collected
depth = []

for quake in cur :
    # Get data from cursor and add to lists or dictionary
    # Dictionaries will hold counts, Lists will append data
    placecounts[quake[4]] = placecounts.get(quake[4], 0) + 1
    regioncounts[quake[7]] = regioncounts.get(quake[7], 0) + 1
    mag.append(quake[1])
    felt.append(quake[6])
#     tsunami.append(quake[]) # Per the last assignment, this was not collected
    depth.append(quake[3])
    
# Print out top earthquake places
print('')
print('Top',howmany,'earthquake places')

x = sorted(placecounts, key=placecounts.get, reverse=True)
for k in x[:howmany]:
    print(k, placecounts[k])
    if placecounts[k] < 10 : break

# Print out top earthquake regions
print('\nTop',howmany,'earthquake regions') 

y = sorted(regioncounts, key=regioncounts.get, reverse=True)
for z in y[:howmany]:
    print(z, regioncounts[z])
    if regioncounts[z] < 10 : break

print("\nBasic Statistics")
# Print out basic statistics like number of earthquakes, mean magnitude, median, standard deviation etc.
print("The total number of earthquakes was: ", len(mag)) 
print("The mean magnitude is: ", np.mean(mag))
print("The median magnitude is: ", np.median(mag))
print("The standard deviation of magnitude is: ", np.std(mag))
print("The biggest earthquake in terms of magnitude is: ", np.amax(mag))

# Convert the "felt/tsunami/depth" list to a NumPy Array
npfilter = np.nan_to_num(np.array((felt,depth),dtype=float))

# Create a NumPy array of Booleans where felt or tsunami" > 0 or depth < 50
filterB = np.array((npfilter[0] > 0,npfilter[1] < 100))

# Create a NumPy array of magnitudes for all of the filtered earthquakes
magF = np.array([mag])

# Print out basic statistics like number of felt earthquakes, mean felt earthquake ... 
print("\nFelt > 0 People Statistics")
print("The total number of earthquakes felt by people is: ", np.sum(npfilter[0][filterB[0]]))
print("The largest amount of people felt for a single earthquake is: ", np.amax(npfilter[0]))
print("The mean of earthquakes felt by people is: ", np.mean(npfilter[0][filterB[0]]))
print("The median of earthquakes felt by people is: ", np.median(npfilter[0][filterB[0]]))
print("The standard deviation of earthquakes felt by people is: ", np.std(npfilter[0][filterB[0]]))

print("\nDepth < 100 Statistics")
print("The sum of all depths for all earthquakes is: ", np.sum(npfilter[1][filterB[1]]))
print("The largest depth of a single earthquake, including those above 100 is: ", np.amax(npfilter[1]))
print("The mean depth of earthquakes is: ", np.mean(npfilter[1][filterB[1]]))
print("The median depth of earthquakes is: ", np.median(npfilter[1][filterB[1]]))
print("The standard deviation depth of earthquakes is: ", np.std(npfilter[1][filterB[1]]))
```
